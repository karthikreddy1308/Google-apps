```python
import pandas as pd
fileapp= '/content/drive/MyDrive/googleplaystore.csv'
fileuserreviews= '/content/drive/MyDrive/googleplaystore_user_reviews.csv'
```

```python
dfapps=pd.read_csv(fileapp)
dfreviews=pd.read_csv(fileuserreviews)
```

```python
# Fill missing 'Rating' values with the mean rating
dfapps['Rating'].fillna(dfapps['Rating'].mean(), inplace=True)

# Drop rows with missing values in 'Type', 'Content Rating', 'Current Ver', and 'Android Ver'
columns_to_drop_na = ['Type', 'Content Rating', 'Current Ver', 'Android Ver']
dfapps.dropna(subset=columns_to_drop_na, inplace=True)

# Check if there are still missing values
print(dfapps.isnull().sum())
```

```python
dfapps.head()
```

```python
print(dfapps.dtypes)
```

```python
# Convert 'Reviews' to integer
dfapps['Reviews'] = dfapps['Reviews'].astype(int)


```

```python
print(dfapps['Installs'].dtype)
```

```python
dfapps['Installs'] = dfapps['Installs'].astype(str)
```

```python
# Remove '+' and ',' and convert 'Installs' to integer
dfapps['Installs'] = (
    dfapps['Installs']
    .str.replace(',', '', regex=False)
    .str.replace('+', '', regex=False)
    .astype(int)
)
```

```python
print(dfapps['Installs'].head())
```

```python
# Remove '$' and convert 'Price' to float
dfapps['Price'] = (
    dfapps['Price']
    .str.replace('$', '', regex=False)
    .astype(float)
)
```

```python
# Function to convert sizes to bytes
def convert_size(size):
    if 'M' in size:
        return float(size.replace('M', '')) * 1e6
    elif 'k' in size:
        return float(size.replace('k', '')) * 1e3
    elif size == 'Varies with device':
        return None  # Handle "Varies with device" as missing value
    else:
        return None  # Handle unexpected values as missing

# Apply the conversion function to 'Size' column
dfapps['Size'] = dfapps['Size'].apply(convert_size)
```

```python
# Check data types and summary statistics
print(dfapps.dtypes)
print(dfapps[['Reviews', 'Installs', 'Price', 'Size']].describe())
```

```python
correlation_matrix = dfapps[['Installs', 'Reviews', 'Rating', 'Price']].corr()

import seaborn as sns
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

Strong Positive Correlation Between Installs and Reviews (0.63):

Apps with more installs tend to have more reviews. This is logical, as a larger user base generally generates more feedback.
Weak Positive Correlation Between Rating and Other Variables (0.039 to 0.054):

Ratings show a minimal positive correlation with installs and reviews, indicating that user satisfaction (as reflected by ratings) is not strongly tied to the popularity of the app or the number of reviews.
No Significant Correlation Between Price and Other Variables:

The near-zero correlations between price and installs (-0.0094), reviews (-0.0076), and ratings (-0.019) suggest that pricing does not significantly affect the number of
 installs, reviews, or user satisfaction.
Installs as a Key Metric:

The moderate positive correlation between installs and reviews highlights installs as a central metric influencing user engagement (via reviews). However, its weak correlation with ratings and price indicates that while an app might be widely downloaded, its quality or cost does not heavily influence its success.

```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
```

```python
# Filter reviews with positive sentiment
positive_reviews = dfreviews[dfreviews['Sentiment'] == 'Positive']['Translated_Review']
```

```python
# Combine all positive reviews into a single string
all_positive_reviews = ' '.join(review for review in positive_reviews if pd.notnull(review))
```

```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt


# Add custom stopwords
custom_stopwords = set(STOPWORDS)
custom_stopwords.update(["app", "apps", "use", "using"])  # Add any words you want to exclude

# Generate the word cloud again, excluding stopwords
wordcloud = WordCloud(
    background_color='white',
    max_words=200,
    width=800,
    height=400,
    colormap='viridis',
    stopwords=custom_stopwords
).generate(all_positive_reviews)

# Display the word cloud
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Common Words in Positive Reviews (Without Stop Words)')
plt.show()

```

This is a word cloud which shows the words that indicate positive reviews. Words like "great", "easy", "love", "best" often indicate why users like these apps

```python
negative_reviews = dfreviews[dfreviews['Sentiment'] == 'Negative']['Translated_Review']
all_negative_reviews = ' '.join(review for review in negative_reviews if pd.notnull(review))
wordcloud_negative = WordCloud(
    background_color='white',
    max_words=200,
    width=800,
    height=400,
    colormap='coolwarm',
    stopwords=custom_stopwords
).generate(all_negative_reviews)

plt.figure(figsize=(12, 8))
plt.imshow(wordcloud_negative, interpolation='bilinear')
plt.axis('off')
plt.title('Common Words in Negative Reviews', fontsize=16)
plt.show()
```

Keywords like 'bad','update','want','hate','frustrating' indicate negative reviews


```python
# Merge apps and reviews datasets
merged_df = pd.merge(dfapps, dfreviews, on='App', how='inner')

# Calculate correlation between Sentiment Polarity and Ratings
correlation = merged_df[['Rating', 'Sentiment_Polarity']].corr()
print(correlation)
```

A correlation of 0.0936 suggests that higher ratings tend to slightly align with more positive sentiment in user reviews, but the relationship is not strong.
This could imply that ratings and review sentiments are influenced by different factors.
<br>1)Bias in Ratings: Some users might give high ratings but leave neutral or even negative comments.
<br>2)Subjectivity in Sentiment Analysis: Sentiment polarity scores are derived from text, which may not fully capture the user's overall sentiment.
Feature Limitations: Ratings may be influenced by factors not reflected in textual reviews, such as app performance, pricing, or specific features.

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.scatterplot(data=merged_df, x='Rating', y='Sentiment_Polarity', alpha=0.5)
plt.title('Ratings vs Sentiment Polarity')
plt.xlabel('Rating')
plt.ylabel('Sentiment Polarity')
plt.show()
```

For most ratings (especially in the range of 3.0 to 5.0), the sentiment polarity values are widely spread across positive, neutral, and negative sentiments.
This indicates that user ratings do not always align with the sentiment of their reviews
<br>There is a noticeable density of points around Rating = 4.0 to 5.0.
These ratings are associated with a mix of sentiments, but the majority seem to fall into the positive or neutral sentiment polarity range.
<br>A significant portion of data points, regardless of rating, have sentiment polarity close to 0.0 (neutral). This suggests that many reviews may not express strong emotions.
